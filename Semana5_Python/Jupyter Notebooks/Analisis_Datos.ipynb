```python
sobreviventes = df['Survived'].value_counts()
print(sobreviventes)
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Age'].dropna(), bins=30, edgecolor='black')
plt.xlabel('Idade')
plt.ylabel('Número de Passageiros')
plt.title('Distribuição de Idades dos Passageiros do Titanic')
plt.show()
```

```python
import seaborn as sns

sns.countplot(x='Pclass', hue='Survived', data=df)
plt.xlabel('Classe')
plt.ylabel('Número de Passageiros')
plt.title('Sobreviventes por Classe')
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.xlabel('Gênero')
plt.ylabel('Número de Passageiros')
plt.title('Sobreviventes por Gênero')
plt.show()
```

```python
import pandas as pd

# Cria uma tabela de contingência entre 'Survived' e 'Sex'
tabela_contingencia = pd.crosstab(df['Survived'], df['Sex'])

# Calcula a correlação
correlacao = tabela_contingencia.corr()

print(correlacao)
```
